```python
# Customer ratings for 5 product categories (scale 1-5) 
customer_A = [4, 5, 2, 3, 4] 
customer_B = [5, 3, 2, 4, 5] 
# Binary preferences (1 = liked, 0 = disliked) 
customer_A_binary = [1, 0, 1, 1, 0, 1] 
customer_B_binary = [1, 1, 1, 0, 0, 1]
```

```python
import pandas as pd
import numpy as np
from numpy.linalg import norm
from scipy.spatial.distance import euclidean, cityblock, hamming
```

```python
customer_A = np.array(customer_A)
customer_B = np.array(customer_B)
customer_A_binary = np.array(customer_A_binary)
customer_B_binary = np.array(customer_B_binary)
```

```python
euclidean_distance = euclidean(customer_A,customer_B)
euclidean_distance
```

```python
manhanttan = cityblock(customer_A_binary,customer_B_binary)
manhanttan
```

```python
cos_sim = np.dot(customer_A, customer_B) / (norm(customer_A) * norm(customer_B))
cos_sim
```

```python
hamming = hamming(customer_A_binary, customer_B_binary)
hamming
```

```python
intersection = np.sum((customer_A_binary == 1) & (customer_B_binary == 1))
union = np.sum((customer_A_binary == 1) | (customer_B_binary == 1))
jaccard_sim = intersection / union
jaccard_sim
```
