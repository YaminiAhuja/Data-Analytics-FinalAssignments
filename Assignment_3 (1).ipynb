```python
selected = df[['Education','BirthDate','Occupation','Gender','MaritalStatus',
                     'HomeOwnerFlag','NumberCarsOwned','NumberChildrenAtHome',
                     'TotalChildren','YearlyIncome','StateProvinceName',
                     'CountryRegionName','BikeBuyer']]
selected
```

```python
selected.info()
```

```python
selected.groupby("BikeBuyer")['YearlyIncome'].mean()
```

```python
(selected.groupby('NumberCarsOwned')['BikeBuyer'].mean() * 100)
```
