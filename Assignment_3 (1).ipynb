```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv("AWCustomers.csv")
```

```python
customers.head(20)
```

```python
customers.info()
```

```python
customers.columns
```

```python
sales = pd.read_csv("AWSales.csv")
sales.info()
```

```python
df = customers.merge(sales,on = 'CustomerID',how  = "inner")
```

```python
df
```

```python
df.info()
```

```python
selected = df[['Education','BirthDate','Occupation','Gender','MaritalStatus',
                     'HomeOwnerFlag','NumberCarsOwned','NumberChildrenAtHome',
                     'TotalChildren','YearlyIncome','StateProvinceName',
                     'CountryRegionName','BikeBuyer']]
selected
```

```python
selected.info()
```

```python
selected.groupby("BikeBuyer")['YearlyIncome'].mean()
selected.groupby("BikeBuyer")['YearlyIncome'].mean().plot(kind='bar')
plt.show()
```

```python
(selected.groupby('NumberCarsOwned')['BikeBuyer'].mean() * 100)
(selected.groupby('NumberCarsOwned')['BikeBuyer'].mean() * 100).plot(kind = "bar")
plt.show()
```

```python
selected['BirthDate'] = pd.to_datetime(selected['BirthDate'])
selected['Age'] = 2024 - selected['BirthDate'].dt.year

selected.groupby('BikeBuyer')['Age'].describe()
```

```python
selected.groupby("MaritalStatus")['BikeBuyer'].mean().plot(kind='bar')
```
